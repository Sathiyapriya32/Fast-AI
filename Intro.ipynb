## Preview Forest Image from Bing download

```python
from fastai.vision.all import *
from pathlib import Path

# Get forest images
forest_path = Path('dataset/forest')
forest_images = list(forest_path.glob('*.jpg')) + list(forest_path.glob('*.png'))

if len(forest_images) > 0:
    print(f"✅ Found {len(forest_images)} forest images")

    # Open and display first forest image
    im_forest = Image.open(forest_images[0])
    display(im_forest.to_thumb(256,256))
else:
    print("❌ No forest images found")
```

```python
from pathlib import Path

searches = 'forest', 'bird'
path = Path('bird_or_not')

if not path.exists():
    path.mkdir(exist_ok=True)

for o in searches:
    dest = (path/o)
    dest.mkdir(exist_ok=True, parents=True)

    # Use Bing downloader instead of DDG
    print(f"📥 Downloading {o} images...")
    downloader.download(
        o,
        limit=50,
        output_dir=str(path),
        adult_filter_off=True,
        force_replace=False,
        timeout=60
    )

    # Resize images
    print(f"🔄 Resizing {o} images...")
    resize_images(dest, max_size=400, dest=dest)

print("✅ Done! Images downloaded and resized.")
```

## Train our model
Some photos might not download correctly which could cause our model training to fail, so we'll remove them

```python
failed = verify_images(get_image_files(path))
failed.map(Path.unlink)
len(failed)
```

## DataBlock
view sample images from it

```python
dls = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=[Resize(192, method='squish')]
).dataloaders(path, bs=32)

dls.show_batch(max_n=6)
```

## fine tuning a pre-trained model (resnet18)

```python
learn = cnn_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(3)
```

## Use our model

```python
is_bird,_,probs = learn.predict(PILImage.create('bird.jpg'))
print(f"This is a: {is_bird}.")
print(f"Probability it's a bird: {probs[0]:.4f}")
```
